```python
%%bash 
#search and display cite names and income data from the source 
cat temp.txt | grep "<title>" | sed 's;<title>\([^,]*\).*;\1;' > cities.txt
cat temp.txt | grep "Estimated median household income" | sed 's;<b>Estimated median household income in 2011:</b> $\([0123456789,]*\).*<br>;\1;' > income.txt
```

```python
#open file to prepare following operate
infile1 = open('cities.txt', 'r')
infile2= open('income.txt', 'r')
```

```python
#put useful data into lists
city=infile1.readlines()
income=infile2.readlines()
ct=[str(line.strip()) for line in city]
icc=[str(line.strip()) for line in income]
print ct, '\n',icc
```

```python
#remove ',' in income data and convert it to intger
count =0
num = 1
hold=''
ic=[]
for icome in icc:
    while count<=5:
        if icome[count]!=',':
            hold = hold+icome[count]   # int(icome[count])*(10**(5-num))
           # print float(icome[count])
            count+=1
            num+=1
        else:
            count+=1
    ic.append(int(hold))
    hold=''
    count=0
    num=1
print ct, '\n', ic
```

```python
from numpy import array
import pandas as pd
from pandas import DataFrame, Series
```

```python
#transfer data into dataframe
income_list = array([ct, ic]).transpose()
income_table = DataFrame(data=income_list, columns=['cities', 'income'])
print income_table
```
