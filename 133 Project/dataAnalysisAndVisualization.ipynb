# Analysis of data and challenges encountered:

Successfully collecting the data, we use linear regression to find the relationship between these factors and average city gas price.




```python
%load_ext rmagic
```

```python
%%R
infile=read.csv("rescale_data.csv")
```

```python
%%R
df=data.frame(infile)
print(df)
cities=df[['cities']]
income=df[['income']]
population_density=df[['population.density']]
automobile_density=df[['automobile.density']]
cost_of_living_index=df[['Cost.of.Living.index']]
gas_price=df[['gas.price']]
```

```python
%%R
coef = lm(df$gas.price ~ df$income+df$population.density+df$automobile.density+df$Cost.of.Living.index)
print(coef)
stats=summary(coef)
print(stats)
```

Here, we encounter the first challenge --- the coefficients for our factors are all not statistically significant, and in fact, they have a 50% chance to be wrong. And thus we need to modifiy our model and dig into the data to see potential outliers and erroneous relationship.

```python
%%R
boxplot(df$gas.price)
```

From the boxplot above, we can see that there are two cities' gas price that's above all the over cities. 

Going to the data, we see that these two cities are San Francisco and Oakland.

And thus we colorcoded San Francisco as 'Blue' and Oakland as 'Red' and plot them in a scatter plot matrix to see whether they are statistical outliers.

```python
%%R
my_colors = c('black', 'black','red', 'black', 'blue', 'black', 'black', 'black', 'black')

point_colors = my_colors2[df$cities]
```

```python
%%R -r 96 -w 800 -h 800
plot(df[c(7,3,4,5,6)],col = point_colors)
```

From the graphs above [especially (gas.price & income),(gas.price & population.density), and (gas.price & cost.of.living.index)], 
we can see that since San Francisco and Oakland have very high gas prices, thus including both of them in the data set will incur substatial standard deviation from the linear regression. But if we exclude Oakland, we can eye ball that the new linear regression will better fits all the points and have smaller SD.
Let's verify the assumption by first plotting the distance of each of these nine cities to the regression line and then compare to the new distance when Oakland is excluded from the data set

```python
%%R

model = lm(df$gas.price ~ df$income) ## model the linear relationship between gas.price and income, for all cities
print(model)
print(summary(model))
```

```python
%%R -r 86 -w 400 -h 300

plot(df$income, df$gas.price)  ##plot all the points and show their distance to the regression line

points(df$income, fitted(model), col='Orange')
segments(df$income, fitted(model), df$income, df$gas.price)
```

```python
%%R

model2 = lm(df$gas.price[-3] ~ df$income[-3])  ## model the linear relationship between gas.price and income, excluding Oakland
print(model2)
print(summary(model2))
```

```python
%%R -r 86 -w 400 -h 300

plot(df$income[-3], df$gas.price[-3])   ##plot all the points and show their distance to the regression line, excluding Oakland

points(df$income[-3], fitted(model2), col='Orange')
segments(df$income[-3], fitted(model2), df$income[-3], df$gas.price[-3])
```

Comparing both graphs, we can see a slight improvement in SD reduction with Oakland excluded. More specifically, if we look at the regreesion results above, we can see that when Oakland is included, SD is 0.0002407 and t-stat is 0.194. But after we exclude Oakland, SD reduces to 0.0002143 and t-stat lowers to 0.12

Thus, we conclude that Oakland is a outlier in the data set, and excluding it significantly improves the estimation.

# Regression Refinement:

SECOND Challenge we encounter is to explain the high t-stat for population density and automobile density, and also take into account the fact that from the plot matrix we had, we see a clear and strong linear relationship between these two factors. (See the graph of population.density and automobile.density)

After researches, we conclude that there is a strong correlation between population density and automobile density and thus including both of them in the regression may incur double counting. Therefore we decided to drop the automobile density. 

Also, we find that the relationship between income and gas price, population density and gas price should be an elasticity. It makes more sense to explain the relationship as "If one city's average income has increased 5%, the local gas price also increased a corresponding percentage).
This means that we cannot simply use lm(gas.price ~ income+population.density+cost.of.living.index), instead we should modify the equation to be log(gas.price) ~ log(income) + log(population.density) + cost.of.living.index)

```python
%%R
new_coef = lm(log(gas_price[-3]) ~ log(income[-3])+log(population_density[-3])+cost_of_living_index[-3]) 
##Remeber to exclude Oakland from the Data Set!!
print(new_coef)
print(summary(new_coef))
```

From this result, we can see that the SD for each factor has reduced significantly and although they are not statistically significant, we managed to reduce the t-stat for income from the previous 0.62690 to the now 0.1865, population density from the previous 0.45155 to the now 0.1710  and the cost of living from the previous 0.59722 to the now 0.2962.

We can see huge improvement of the regression result by firstly excluding Oakland, secondly dropping Automobile_density, and thirdly adjusting the regression function.

# Conclusion:
